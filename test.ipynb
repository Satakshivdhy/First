```python
n = int(input())

arr = list(map(int,input().split()))
print(arr)
sorted_arr = sorted(arr)
print(sorted_arr)
rank_map = {}
rank = 1
for num in sorted_arr:
    if num not in rank_map:
        rank_map[num] = rank
        rank += 1
for num in arr:
    print(rank_map[num], end=" ")
print(rank_map)
```

```python
def replaceWithRank(arr):
    rank_map={}
    rank = 1
    for num in sorted(arr):
        if num not in rank_map:
            rank_map[num]=rank_map
            rank+=1
    return[rank_map[num] for num in arr]
n=int(input())
arr = list(map(int, input().split()))
result = replaceWithRank(arr)
print(*result)
```

```python
n = int(input())
arr = list(map(int,input().split()))
print(arr)
freq={}
result=[]
for num in arr:
    freq[num] = freq.get(num , 0)+1
print(freq)
for num in arr:
    if freq[num]==1:
        result.append(num)
if result:
    print(result)
else:
    print(-1)
 
```
